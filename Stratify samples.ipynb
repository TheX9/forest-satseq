```python
def calc_ratio(gdf):
    return pd.merge(gdf, df_stats, on='point_id')['loss_sum_2019'].sum()/(len(gdf))
```

```python
def remove_tiles(gdf, amount, lower = None, higher = None):
    gdf_merged = pd.merge(gdf, df_stats, on='point_id')
    if lower is not None:
        remove_index = gdf_merged[gdf_merged['loss_sum_2019']>=lower].sample(n=amount, random_state = 0)
    if higher is not None:
        remove_index = gdf_merged[gdf_merged['loss_sum_2019']<=higher].sample(n=amount, random_state = 0)
    if lower is not None and higher is not None:
        remove_index = gdf_merged[(gdf_merged['loss_sum_2019']>=lower)&(gdf_merged['loss_sum_2019']<=higher)].sample(n=amount, random_state = 0)
    
    remove_index = list(remove_index['FID'])
    gdf = gdf[~gdf['FID'].isin(remove_index)]
    
    return gdf

def remove_tiles_by_random_sampling(gdf, amount, target_ratio=None, gap=None, fixed_random_state = None, random_state = 0):
    best_random_state = 0
    best_ratio = 0
    best_gap = 1000
    if fixed_random_state:
        gdf_stratified = gdf.sample(n=amount, random_state = fixed_random_state)
        random_state = fixed_random_state
    else:
        gdf_stratified = gdf.sample(n=amount, random_state = random_state) 
        while abs(target_ratio-calc_ratio(gdf_stratified)) > gap or target_ratio > calc_ratio(gdf_stratified):
            random_state += 1
            gdf_stratified = gdf.sample(n=amount, random_state = random_state) 
            new_ratio = calc_ratio(gdf_stratified)
            new_gap = abs(new_ratio-target_ratio)
            if new_gap <= best_gap:
                best_ratio = new_ratio
                best_random_state = random_state
                best_gap = new_gap
                print(best_random_state,best_ratio)
                
            if random_state > 100000:
                break
           
            
                
                
    return random_state, gdf_stratified, calc_ratio(gdf_stratified)
    
```

```python
import re
# Get unallowed points
with open('O:/2013_2019/process_samples_baseline_True.log', 'r') as file:
    data = file.read()
    
# Get tf record level
removable_sub_dirs = {}
entries = re.findall('(?<=Features)(.*)(?=//satseq_sampleRegions_baseline_0)',str(data))
for entry in entries:
    features = re.findall('(?<=\')(.*)(?=\')', str(entry))
    features = [feature.replace("'","") for feature in features]
    features = features[0].split(',')
    sub_dir = re.findall('(?<=are missing for data/raw_data/baseline/2013_2019_preprocessed/)(.*)',str(entry))
    removable_sub_dirs[sub_dir[0]] = features

# Get point level
sub_strs = re.findall('(?<=Missing values for feature )(.*)(?=\n)',data)

missing_point_dict = {}
for sub_str in sub_strs:
    point = re.search('(?<=point )(.*)',sub_str).group(1)
    feature = re.search('(.*)(?= for)',sub_str).group(1)
    if point not in missing_point_dict:      
        missing_point_dict[point] = []
    missing_point_dict[point].append(feature)
```

```python
unallowed_points = [missing_point.replace("_2013_2019","") for missing_point in missing_point_dict]
```

```python
gdf_stratified_train = gpd.read_file(gdf_basefile_name%('stratified_train'))
gdf_stratified_test = gpd.read_file(gdf_basefile_name%('stratified_test'))
gdf_stratified_val = gpd.read_file(gdf_basefile_name%('stratified_val'))
```

```python
# Find best stratified sample split for data
def find_best_split(gdf, percent, gap = None, fixed_random_state = None, target_ratio = None, random_state = 0):
    best_gap = 10000
    best_random_state = 0
    best_ratio = 0
    amount = int(percent*len(gdf))
    gdf = gdf[~gdf['point_id'].isin(unallowed_points)]
    
    
    if fixed_random_state:
        random_state = fixed_random_state
        gdf_stratified = gdf.sample(n=amount, random_state = random_state) 
    else:
        if not target_ratio:
            target_ratio = calc_ratio(gdf)
        gdf_stratified = gdf.sample(n=amount, random_state = random_state) 
        best_gap = abs(target_ratio-calc_ratio(gdf_stratified))
        
        while abs(target_ratio-calc_ratio(gdf_stratified)) >= gap:
            random_state += 1
            gdf_stratified = gdf.sample(n=amount, random_state = random_state) 
            
            new_ratio = calc_ratio(gdf_stratified)
            new_gap = abs(new_ratio-target_ratio)
            if new_gap <= best_gap:
                best_ratio = new_ratio
                best_random_state = random_state
                best_gap = new_gap
                print(best_random_state,best_ratio)
            
            if random_state > 10000:
                break
                
    return random_state, gdf_stratified, calc_ratio(gdf_stratified)
    
```

```python
# Find best sampling to get a stratified baselie split 1%
(random_state_train,
gdf_stratified_train_baseline,
ratio_train) =  find_best_split(gdf_stratified_train,
                                percent = 0.01,
                                gap = 0.01,
                                target_ratio = 600, 
                                random_state = 4269)
                                
print(random_state_train, ratio_train)

(random_state_val,
gdf_stratified_val_baseline,
ratio_val) =  find_best_split(gdf_stratified_val,
                              percent = 0.01,
                              gap = 0.01,
                              target_ratio = 600,
                              random_state = 1337)
print(random_state_val, ratio_val)

(random_state_test,
gdf_stratified_test_baseline,
ratio_test) =  find_best_split(gdf_stratified_test,
                               percent = 0.01,
                               gap = 0.01,
                               target_ratio = 600)
print(random_state_test, ratio_test)
```
